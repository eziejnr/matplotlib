```python
import matplotlib.pyplot as plt 
import numpy as np


landuse =("industrial", "agriculture", "commercial", "residential", "water")
landusedata ={
    "2000": np.array([15000,20000, 10000,5000,4321]),
    "2010": np.array([17000,10000, 16000, 8000, 3321]),
    "2020": np.array([15000,8000, 20000, 7000, 4321])
}
width=0.7
fig,ax=plt.subplots()
bottom=np.zeros(5)
for year, data in landusedata.items():
    Plot=ax.bar(landuse,data,width,label=year,bottom=bottom)
    ax.bar_label(Plot,label_type="center")
    bottom+=data

plt.legend()
ax.set_ylabel("area")
ax.set_xlabel("landuse landcover")
ax.set_title("landuse landcover change analysis")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
np.random.seed(123321)
LandUse = ("Forest", "Agriculture", "Urban", "Water", "Grassland")
y_pos = np.arange(len(LandUse))
changes = 4 + 6 * np.random.rand(len(LandUse))
error = np.random.rand(len(LandUse))
fig, ax = plt.subplots()
hbar = ax.barh(y_pos, changes, xerr=error, align="center")
ax.set_yticks(y_pos, labels=LandUse)
ax.invert_yaxis()
ax.set_title("Land Use Changes Over Time")
ax.set_xlabel("Changes in Land Use") 
ax.set_ylabel("Land Use Categories")
ax.bar_label(hbar, labels=[f'{c:.3f}%' for c in changes])
ax.set_xlim(right = 12)
plt.figtext(0.9, 0.01, "Data source: https://data.apps.fao.org/", ha="right", fontsize=6)
ax.set_xlabel("Changes in Land Use and Cover", ha = "right")
plt.show()
```
