```python
import matplotlib.pyplot as plt
import numpy as np

fig, ax = plt.subplots()
landuse = ["water/meshland", "urban", "shrubs", "forest", "baresoil"]
values = [3.083, 28.321, 16.077, 16.762, 35.757]
labels = ["water/meshland", "urban", "shrubs", "forest", "baresoil"]
colors = ["tab:cyan", "tab:blue", "tab:orange", "tab:green", "tab:red"]

ax.bar(landuse, values, label=labels, color=colors)
ax.set_ylabel("year1993")
ax.set_xlabel("landuse landcover")
ax.legend(title="LULC", loc=(2) 
ax.set_title("Land Cover Land Use Class Changes Year 1993") 

plt.show()

```
